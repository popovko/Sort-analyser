```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showBarPlotForSorts(
    sorts,
    value,
    dataType,
    shapeArray
):
    y_pos = np.arange(len(sorts))
    plt.bar(y_pos, value, align='center', alpha=0.5)
    plt.xticks(y_pos, sorts)
    plt.ylabel('Time miliseconds')
    plt.title('Data type = ' + dataType + 'shape = ' +shapeArray)
    plt.show()
```

```python
sorts = ['Library', 'Counting', 'Smooth', 'Intro', 'stl::sort', 'stl::stable_sort']
```
